```python
import scipy.stats as sps
import numpy as np
import matplotlib.pyplot as plt
import seaborn as sns
%matplotlib inline

#класс генерирующий ф.р.
class cubic_gen(sps.rv_continuous):
    def _pdf(self, x):
        return 4 * x ** 2 / 15

cubic = cubic_gen(a=1, b=2, name='cubic')
sample = cubic.rvs(size=1000)

#посторим гистограмму нашего распределния
plt.hist(sample,10)
plt.xlabel('интервал')
plt.ylabel('кол-во вхожедений в каждый карман')
```

```python
x = np.linspace(1,2,1000)
plt.plot(x, cubic.pdf(x), color='red', label='Плотность случайной величины')

plt.hist(sample, density=True, label='Гистограмма случайной величины') 

plt.legend(fontsize=14, loc=1)
plt.xlabel('интервал')
plt.ylabel('')
```

```python
"""
Ваша задача — оценить распределение выборочного среднего вашей случайной величины при разных объёмах выборок. 
Для этого при трёх и более значениях n (например, 5, 10, 50) 
сгенерируйте 1000 выборок объёма n и постройте гистограммы распределений их выборочных средних. Реализуем это!
"""
n = [1,5,10]
kek = np.zeros(1000)
for i in np.arange(1000):
    sample = cubic.rvs(size=n[2])
    kek[i] = sample.mean()
plt.hist(kek)
plt.ylabel('fraction of samples')
plt.xlabel('x')
var = np.var(kek)
```

```python
print('среднее = {}'.format(kek.mean()))
print('дисперсия = {}'.format(var))
print('среднеквадртаичное отклонение = {}'.format(kek.std()))
```

```python
norm = sps.norm(kek.mean(),np.sqrt(var))
plt.hist(kek,density=True, label='Гистограмма')
plt.plot(x,norm.pdf(x),label='плотность')
plt.ylabel('fraction of samples')
plt.xlabel('x')
plt.legend(fontsize=14, loc=1)
```
