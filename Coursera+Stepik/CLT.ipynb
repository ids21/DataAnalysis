```python
import pandas as pd
import numpy as np
import matplotlib.pyplot as plt
import scipy.stats as sps
from scipy.stats import expon
import math
%matplotlib inline
```

Функция формирования гистограммы распределения выборочного среднего и графика плотности соответствующего нормального распределения
$\mu = 1$

$\sigma^2 = 1$

Поэтому согласно ЦПТ нормальное распределение будет следующим:

 $\mathcal{N}(1,1/n)$

```python
# n - объем выборки 
def exp(n):
    #генерация выборки экспоненциального распределения
    sample = sps.expon.rvs(size=(1000, n))
    #вычисление выборочного среднего
    means = sample.mean(axis=1)
    plt.figure(figsize=(6, 4))
    #построение гистограммы
    plt.hist(means, density=True, bins=20, alpha=0.7)
    #создание 1000 точек
    x = np.linspace(0, 2, 1000)
    plt.plot(x, sps.norm(loc=1, scale=np.sqrt(1./n)).pdf(x), label='theoretical density')
    plt.xlabel('sample mean')
    plt.ylabel('density')
    plt.title('Distribution of sample means: $\mathcal{N}(1, ' + str(round(1. / n, 3)) + ')$')
    plt.legend()
    plt.show()
```

Построение графиков

```python
exp(5)
exp(10)
exp(50)
```

# Вывод

Из графиков заметим, что выборочное среднее с увеличением объема выборки приближается к математическому ожиданию нормального распределения, а это согласуется с ЦПТ.
