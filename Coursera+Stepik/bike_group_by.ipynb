```python
df.starttime = pd.to_datetime(df['starttime'])
df.stoptime = pd.to_datetime(df['stoptime'])
dt = df[(df.starttime.dt.hour>=18) & (df.stoptime.dt.hour <=20)].groupby('start station id').agg({'start station id':'count'})
dt.rename(columns={'start station id': 'count'}, inplace=True)
dt.sort_values('count')
```
